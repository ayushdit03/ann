```python
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2, 100)
```

```python
def linear(x):
    return x

def sigmoid(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return np.maximum(0,x)

def tanh(x):
    return np.tanh(x)
```

```python
sbn.lineplot(x=x, y=linear(x), label="linear")
plt.grid(True)
plt.title("Linear")
plt.show()
sbn.lineplot(x=x, y=sigmoid(x), label="sigmoid")
plt.grid(True)
plt.title("sigmoid")
plt.show()
sbn.lineplot(x=x, y=relu(x), label="relu")
plt.grid(True)
plt.title("relu")
plt.show()
sbn.lineplot(x=x, y=tanh(x), label="tanh")
plt.grid(True)
plt.title("tanh")
plt.show()
```
